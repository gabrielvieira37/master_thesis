```python
import scipy.io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_mat(mat, key):
    mdata = mat[key]
    
    mdtype = mdata.dtype

    # Use ndata to create a dict and access it using
    # respective key to create a dataframe
    ndata = {n: mdata[n][0, 0] for n in mdtype.names}
    return ndata
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
enterp = pd.DataFrame(ndata['MonthlyEnterprise'], columns=names).fillna(method='bfill').fillna(0)
enterp.describe()
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
debt = pd.DataFrame(ndata['MonthlyDebt'], columns=names).fillna(method='bfill').fillna(0)
debt.describe()
```

```python
book_to_mkt_ratio = (enterp-debt)/mcap
book_to_mkt_ratio.describe()
```

```python
two_third_size = int((book_to_mkt_ratio['VALE3'].describe()['count'])*(2/3))
train = book_to_mkt_ratio['VALE3'][:two_third_size]
validation = book_to_mkt_ratio['VALE3'][two_third_size:]
```

```python
plt.figure(figsize=(12,8))
plt.title("Vale Daily book to market ratio")
plt.ticklabel_format(style='plain')
train.plot(c='blue', label='train')
validation.plot(c='orange', label='validation')
plt.legend()
# plt.grid()
plt.show()
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
splits = TimeSeriesSplit(n_splits=3)
```

```python
book_to_mkt_ratio
```

```python
for train, val in splits.split(book_to_mkt_ratio):
    train_X = book_to_mkt_ratio.loc[train]
    print(train_X.shape)
    val_X = book_to_mkt_ratio.loc[val]
    print(val_X.shape)
    print()
    plt.figure(figsize=(12,8))
    plt.title("Vale Daily book to market ratio")
    plt.ticklabel_format(style='plain')
    train_X['VALE3'].plot(c='blue', label='train')
    val_X['VALE3'].plot(c='orange', label='validation')
    plt.legend()
    # plt.grid()
    plt.show()
```
